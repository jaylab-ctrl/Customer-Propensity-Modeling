```python
%%writefile app.py

import streamlit as st
from langchain.llms import HuggingFaceHub
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

st.title("Cause, Effect and Reccomendation for your Customers")

template_ip = """You are a marketing strategist for a well-reputed company and have been making strategies for 20+ years. \n
There is a case of customer identification number {customer_id}, who is currently a customer at the company. The customer has a rating of {rating} (1 for lowest and 5 for highest),\n
a churn rate of {churn_rate} (0.01 for lowest and 0.99 for highest), and an engagement rating of {engage_rating} (1 for new customer, 2 for low spender, 3 for loyal customer, and 4 for big spender).\n
So, tell me what the causes and effects this company will face in the above scenario are, and what recommendation you give to a company so that it will retain this specific customer."""

fin_prompt = PromptTemplate(input_variables = ['customer_id', 'rating', 'churn_rate', 'engage_rating'],
                          template = template_ip)

cust_id = st.text_input("Enter the customer_id: ")
sat_rating = st.text_input("Enter customer rating: ")
churn = st.text_input("Enter the churn rate: ")
engage = st.text_input("Enter the engagement rating: ")

fin_llm = HuggingFaceHub(repo_id="declare-lab/flan-alpaca-large", model_kwargs={"temperature":1e-10})

fin_chain = LLMChain(llm = fin_llm, prompt = fin_prompt)

if st.button("Generate"):
  response = fin_chain.run({
    'customer_id' : 'cust_id',
    'rating' : 'sat_rating',
    'churn_rate' : 'churn',
    'engage_rating' : 'engage'
  })
  print(response)

```
